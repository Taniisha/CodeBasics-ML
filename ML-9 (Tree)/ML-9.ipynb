```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("salary.csv")
```

```python
df
```

```python
# we can see  company, job, degree have text data, so we have to convert it into dummy variables.  
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# since we have three coumns therefore we have to make three objects of LabelEncoder class.
```

```python
inputs = df.drop('salary_more_then_100k', axis='columns') 
```

```python
target = df.salary_more_then_100k
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
```

```python
 inputs['company_n']
```

```python
inputs['job_n'] = le_company.fit_transform(inputs['job'])
```

```python
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['company','job','degree'], axis='columns')
```

```python
inputs_n  #our final x
```

```python
target      #this is our final y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs_n, target, test_size=0.2)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
model.predict([[0, 1,0]])
```
