```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("eight.csv")
```

```python
df
```

```python
plt.scatter(df.age, df.insurance, color='red', marker='+')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['age']], df.insurance, test_size=0.1)
```

```python
x_test
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.predict_proba(x_test)     #the result is giving two classes, 1st when the person is not buying insurance and second when he buys insurance 
```

model.predict(56)

```python
model.predict([[25]])
```
